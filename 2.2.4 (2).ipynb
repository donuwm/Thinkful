```python
# Import data
df = pd.read_csv('BTC_USD.csv')
```

```python
print(df.head())
```

```python
# Plot the closing price in USD
plt.plot(df['Close'])
plt.ylabel('Closing price in USD')
plt.xlabel('days')
plt.show()
```

### The above graph is the bitcoin closing price from August 19, 2008 to November 20, 2008.  Y axis represents the price in dollars and the X axis represents days beginning on Aug 19.

```python
plt.plot(df['High'], color='red')
plt.show()
```

###  The above graph shows the highest Bitcoin price of the day from Aug 19 - Nov 20 of 2018 in the color red. 

```python
df_difference = df['High'] - df['Low']
```

```python
plt.plot(df_difference, color='purple')
```

```python
plt.show()
```

### The above graph shows the difference of the Bitcoin high and low price of the day from August 19 to November 20, 2018.

```python
plt.plot(df['Volume'])
```

```python
plt.show()
```

### The above graph shows the Bitcoin volume for each day from 8/19/2008 - 11/20/2008.  Y axis is in billions of Bitcoins traded; X axis is the 24 hour period starting on 8/19.

```python
plt.figure(figsize=(10, 5))

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.subplot(2, 1, 1)
plt.plot(df_difference, color='yellow')
plt.ylabel('Diff in USD')
plt.title('High/Low Difference')

plt.subplots_adjust(hspace=.5)

plt.subplot(2, 1, 2)
plt.plot(df['Volume'], color='green')
plt.ylabel('Bitcoins Traded')
plt.title('Volume per day')
plt.show()
```

### The above graph shows the diference in the high and low price for each trading day (top graph) versus the volume of bitcoins traded each trading day (bottom graph).  The daily price range is highly correlated with the daily volume.
