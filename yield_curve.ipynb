# What is a yield curve ?  

The yield curve is the line plot of Treasure bonds (or related debt instrument) vs the yield of those bonds.  The curve becomes inverted - the slope of the line is negative - when long term rates are lower than short term rates.  Why would this happen ?  If an investor believes that the long term prospects of alternative investments such as the stock market are overvalued, then she may decide to invest in a very safe long term bond.  If enough investors have this same opinion, then the yield curve inverts.  

See the current yield curve below - as of August, 2019:

```python
# import necessary modules
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# create the datafram and produce the line chart
df = pd.DataFrame({
    'yield': [2.05, 1.96, 1.94, 1.89, 1.72, 1.50, 1.44, 1.42, 1.49, 1.55, 1.84, 2.04]
}, index=['1m', '2m', '3m', '6m', '1y', '2y', '3y', '5y', '7y', '10y', '20y', '30y'])
lines = df.plot.line(figsize=(10, 6))
lines.set_xlabel("Term")
lines.set_ylabel("Yield")
```

The above plot shows how the **yield curve is inverted from the get-go**.  It doesn't begin to take it's more normal, positive sloping shape until the change from the 5 year bond to the 7 year bond.

Next, reverting back to the inverted yield curve, the following plot shows the difference between the 10 year bond and the 3 month government note.  These are the two common debt instruments to compare for a looming recession.

```python
# read in all of the local csv files into pandas:
df = pd.read_csv('10y_3m.csv')

# Convert price to a number, with coersion to throw NA's if conversion not possible
df['price'] = pd.to_numeric(df['T10Y3M'], errors='coerce')

# Remove NA's
df.dropna(inplace=True)
```

```python
# convert index to datetime and produce pyplot
df.index = pd.to_datetime(df.DATE)
df.plot(figsize=(12, 8))
plt.title('10 Year Treasury minus 3 Month Treasury')
plt.ylabel('Percent')

# print horizontal line at 0
plt.axhline(y=0, color='black')

# print vertical lines to start and stop periods of recession
plt.axvline(x='1990-07-01', color='green')
plt.axvline(x='1991-03-01', color='red')

plt.axvline(x='2001-03-01', color='green')
plt.axvline(x='2001-11-01', color='red')

plt.axvline(x='2007-12-01', color='green')
plt.axvline(x='2009-06-01', color='red')

plt.show()
```

The difference in the 10 yr vs 3 month note was negative prior to the official start of the last 3 recessions.  As shown above, the difference is currently negative - as of August 22, 2019.  The green vertical line marks the start of the recession while the red vertical line marks the end.  It looks like this chart is an extremely relevant predictor of recessions.  Definition of recession:  a period of temporary economic decline during which trade and industrial activity are reduced, generally identified by a fall in GDP in two successive quarters.
