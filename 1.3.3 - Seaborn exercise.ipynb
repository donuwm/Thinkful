```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### 1.  Choose one variable and plot that variable four different ways.

```python
# Import data into df_usd data frame
df_usd = pd.read_csv('btc_usd_sep2_nov30.csv')

# print 1st 5 rows of df_usd (spot price)
df_usd.head()
```

```python
print('Scatterplot of spot price')
plt.scatter(df_usd['Date'], df_usd['High'])
#plt.scatter({'Sep', 'Oct', 'Nov'}), df_usd['High']
plt.xlabel('Date from Sep2 to Nov30, 2018')
plt.ylabel('Price in USD')
plt.rcParams['figure.figsize'] = (10.0, 5.0) 
plt.rcParams.update({'font.size': 12})
plt.show()

# Summary:
# This plot works fairly well in giving the brief overview of the individual price points.  
# You can also see signigicant daily changes in price without too much trouble.
```

```python
print('Histogram of bitcoin spot price from Sep2 to Nov30, 2018')
plt.hist(df_usd['High'])
plt.xlabel('Price in USD')
plt.ylabel('Percent in each bin')
plt.rcParams['figure.figsize'] = (10.0, 5.0) 
plt.rcParams.update({'font.size': 12})
plt.show()

# Summary:
# This also works well to see the overall distibution.  Date of the price movement is not present here, 
# but we can cleary see where most of the price points fall.
```

```python
print('QQ plot of bitcoin spot price from Sep2 to Nov30, 2018')
norm = np.random.normal(0, 1, len(df_usd['High']))
norm.sort()
plt.plot(norm, df_usd['High'].sort_values(), "o") 
plt.xlabel('Mean value of a sorted normal distribution')
plt.ylabel('Price in USD')
plt.rcParams['figure.figsize'] = (10.0, 5.0) 
plt.rcParams.update({'font.size': 12})
plt.show() 

# Summary:
# This shows that the distribution does not clearly fit a normal distribution as most 
# of the price points fall in around $6500.
```

```python
print('Boxplot of bitcoin price in USD')
plt.boxplot(df_usd['High'])
plt.xlabel('Bitcoin spot price disribution')
plt.ylabel('Price in USD')
plt.rcParams['figure.figsize'] = (10.0, 5.0) 
#plt.rcParams['font.size'] = 30
plt.show()

# Summary:
# Boxplot does not work very well here because there are too many "outliers" and the data is too
# concentrated within a specific price range at the $6500 level.
```

### 2.  Choose two continuous variables, and plot them three different ways.


```python
df_plant_height = pd.read_csv('plant_height.csv')
df_plant_height.head(n=50)
```

```python
sns.scatterplot(x='height', y='temp', data=df_plant_height)
plt.xlabel('Height (m)')
plt.ylabel('Temperature (C)')
plt.rcParams['figure.figsize'] = (10.0, 5.0) 
plt.show()

# Summary:
# This is somewhat helpful as you can see a slight increase in plant height as the temperature gets higher; 
# you have to study and strain a little, however. 
```

```python
sns.scatterplot(y='height', x='temp', data=df_plant_height)
plt.xlabel('Height (m)')
plt.ylabel('Temperature (C)')
plt.rcParams['figure.figsize'] = (10.0, 5.0) 
plt.show()

# This one is better; you can more clearly see higher plant heights at higher temps.
```

```python
sns.lmplot(y='height', x='temp', data=df_plant_height)
plt.xlabel('Height (m)')
plt.ylabel('Temperature (C)')
plt.rcParams['figure.figsize'] = (10.0, 5.0) 
plt.show();

# Summary:
# This plot is better still which shows a regression line gving a mathmatical relationship between 
# plant height and temperature.  You can see how it would be easier to pridict one of the variables
# when the other is known.
```

```python
sns.regplot(y='height', x='temp', data=df_plant_height)
plt.xlabel('Height (m)')
plt.ylabel('Temperature (C)')
plt.rcParams['figure.figsize'] = (10.0, 10.0) 
plt.show()

# Summary:
# This is the best graph; very similar to the lmplot, but in a better format.
# Regplot accepts more data formats for x and y variables, but has a more limited
# set of features compared to lmplot.
```

### 3.  Choose one continuous variable and one categorical variable, and plot them six different ways.

```python
sns.set(style="whitegrid")
plt.xlabel('Growthform')
plt.ylabel('Height (m)')
plt.rcParams['figure.figsize'] = (10.0, 5.0) 
sns.barplot(x='growthform', y='height', data=df_plant_height)
plt.show()

# Summary:
# A bar plot represents an estimate of central tendency for a numeric variable with the height of each 
# rectangle and provides some indication of the uncertainty around that estimate using error bars.
# This is an excellent way to show the differences in height of the various growthforms.
```

```python
sns.set(style="whitegrid")
plt.xlabel('Height (m)')
plt.ylabel('Growthform')
plt.rcParams['figure.figsize'] = (10.0, 5.0) 
sns.barplot(y='growthform', x='height', data=df_plant_height)
plt.show()

# Summary:
# Just a rotated view of the graph above; probably not quite as easy to read.
```

```python
sns.set(style="darkgrid")
plt.xlabel('Height (m)')
plt.ylabel('Growthform')
plt.rcParams['figure.figsize'] = (10.0, 5.0) 
sns.pointplot(x='growthform', y='height', data=df_plant_height)
plt.show()

# Summary:
# This pointplot shows the average heights for all of the growthforms with error bars.
# Probably not quite as helpful as the barplot.
```

```python
sns.set(style="darkgrid")
plt.xlabel('Height (m)')
plt.ylabel('Growthform')
plt.rcParams['figure.figsize'] = (10.0, 5.0) 
sns.pointplot(x='growthform', y='height', data=df_plant_height, join=False)
plt.show()

# Summary:
# Same graph as above without joining the growthforms.  Not very exciting.
```

```python
sns.set(style="ticks")
#plt.xlabel('Height (m)')
#plt.ylabel('Growthform')
plt.rcParams['figure.figsize'] = (20.0, 10.0) 
sns.catplot(x='growthform', y='height', data=df_plant_height)
plt.show()

# Summary:
# This is very interesting because it shows the number of plants in each category.
```

```python
sns.set(style="ticks")
#plt.xlabel('Height (m)')
#plt.ylabel('Growthform')
plt.rcParams['figure.figsize'] = (20.0, 10.0) 
sns.catplot(x='growthform', y='height', hue='hemisphere', data=df_plant_height)
plt.show()

# Summary:
# This is a little more interesting showing the hemisphere for each plant with a legend.
```
